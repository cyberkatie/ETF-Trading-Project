# Coal
* production

```python
url = 'https://api.eia.gov/v2/coal/aggregate-production/data/?api_key=' + eia_key +'&frequency=annual&data[0]=production&sort[0][column]=period&sort[0][direction]=desc&offset=0&length=5000'
```

```python
response = requests.get(url).json()
response
```

```python
#putting data from every entry into lists and then into a dataframe 
state_name = []
coal_production = []
mine_type = []
coal_rank_id = []
period = []
for i in range(len(response['response']['data'])):

    state_name.append(response['response']['data'][i]['stateRegionId'])
    coal_production.append(response['response']['data'][i]['production'])
    mine_type.append(response['response']['data'][i]['mineTypeId'])
    coal_rank_id.append(response['response']['data'][i]['coalRankId'])
    period.append(response['response']['data'][i]['period'])
    
```

```python
coalprod = {'State Region Id' : state_name,
       'Coal Production (short tons)' : coal_production,
       'Mine Type' : mine_type,
       'Coal Rank ID' : coal_rank_id,
       'Period' : period}
coal_production_df = pd.DataFrame(coalprod)
coal_production_df
```

```python
#need to sort data
prod_UND = coal_production_df.loc[(coal_production_df['Mine Type'] == 'UND') & (coal_production_df['Coal Rank ID'] == 'BIT')]
```

```python
plt.scatter(prod_UND['Period'], prod_UND['Coal Production (short tons)'])
```

* pricing (market price)

```python
url = 'https://api.eia.gov/v2/coal/market-sales-price/data/?api_key=' + eia_key+'&frequency=annual&data[0]=price&sort[0][column]=period&sort[0][direction]=desc&offset=0&length=5000'
```

```python
response = requests.get(url).json()
response
```

```python
region = []
price = []
market_type = []
period = []
for i in range(len(response['response']['data'])):

    region.append(response['response']['data'][i]['stateRegionId'])
    price.append(response['response']['data'][i]['price'])
    market_type.append(response['response']['data'][i]['marketTypeId'])
    period.append(response['response']['data'][i]['period'])
    
```

```python
coalprice = {'Region' : region,
            'Price (Dollars per short ton)' : price,
            'Market Type' : market_type,
            'Period' : period}
coal_price_df = pd.DataFrame(coalprice)
coal_price_df
```

```python
coal_price_df.drop(coal_price_df[coal_price_df['Price (Dollars per short ton)'] == 'w'].index, inplace = True)
```

```python
coal_price_df
```

```python
#sort for the total market price for the appalachia region
app_price = coal_price_df.loc[(coal_price_df['Market Type'] == 'TOT') & (coal_price_df['Region'] == 'APP')]
```

```python
plt.plot(app_price['Period'], app_price['Price (Dollars per short ton)'])
plt.xlabel('Period (Years)')
plt.ylabel('Coal Price ($ per Short Ton)')
plt.title('Coal Price from 2000 to 2021 in the Appalachia Region')
plt.show()
```

```python
#we can chart other regions too if we want
ak_price = coal_price_df.loc[(coal_price_df['Market Type'] == 'TOT') & (coal_price_df['Region'] == 'AK')]
ino_price = coal_price_df.loc[(coal_price_df['Market Type'] == 'TOT') & (coal_price_df['Region'] == 'INO')]
```

```python
plt.plot(ak_price['Period'], ak_price['Price (Dollars per short ton)'], color = 'red', label = 'alaska')
plt.plot(ino_price['Period'], ino_price['Price (Dollars per short ton)'], color = 'blue', label = 'INO')
plt.legend()
```
