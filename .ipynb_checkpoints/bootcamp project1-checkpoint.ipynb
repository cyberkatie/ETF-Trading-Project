```python
data = []
state_name = []
fuel = []
carbon_emissions = []
sector_id = []
period = []
for i in range(len(response['response']['data'])):

    state_name.append(response['response']['data'][i]['state-name'])
    fuel.append(response['response']['data'][i]['fuel-name'])
    carbon_emissions.append(response['response']['data'][i]['value'])
    sector_id.append(response['response']['data'][i]['sectorId'])
    period.append(response['response']['data'][i]['period'])
    
```

```python
state_name
```

```python
data = {'State' : state_name,
       'Fuel' : fuel,
       'Carbon Emissions' : carbon_emissions,
       'Sector ID' : sector_id,
       'Period' : period}
```

```python
df = pd.DataFrame(data)
df
```
